## Task 1: Data Import and Merging

The first step of this analysis is to import the provided datasets and combine them into a single dataset for further processing. Two datasets were provided, each containing related information about customers and their financial attributes.

Both datasets were loaded into the Python environment using the pandas library. Since the datasets share a common identifier, they were merged using the "ID" column as the primary key.

Merging the datasets ensures that all relevant information is available in one place, which is essential for accurate analysis and model building. After merging, the dataset was checked to confirm that the operation was successful and that no duplicate records were introduced.

```python
#Load Data
import pandas as pd
df1 = pd.read_csv("Credit_Card_Dataset_2025_Sept_1.csv")
df2 = pd.read_csv("Credit_Card_Dataset_2025_Sept_2.csv")

print("Dataset 1 shape:", df1.shape)
print("Dataset 2 shape:", df2.shape)
```

```python

print(df1.columns)
print(df2.columns)
```

```python
df1 = df1.drop(columns=['Unnamed: 0'])
```

An unnecessary column named "Unnamed: 0" was identified in the dataset. This column was likely created during the data export process and does not provide any meaningful information for analysis.

Therefore, it was removed to improve data quality and ensure that only relevant features are used in further analysis and model building.

```python
len(set(df1['ID']).intersection(set(df2['User'])))
```

To ensure that the datasets can be merged correctly, the common values between the key columns were examined. The intersection between the ID column in the first dataset and the User column in the second dataset was calculated.

This step helps to confirm whether both datasets share matching records and ensures that the merge operation will be meaningful.

```python
#merge Data
merged_df = pd.merge(df1, df2, left_on='ID', right_on='User', how='inner')
```

The datasets were merged using the "ID" column from the first dataset and the "User" column from the second dataset. Since the column names differ, the merge was performed using the left_on and right_on parameters.

An inner join was used to ensure that only matching records from both datasets are included in the final dataset. This helps to maintain data consistency and avoids introducing missing values from unmatched records.

```python
print("Merged dataset shape:", merged_df.shape)
```

```python
print("Duplicate IDs:", merged_df['ID'].duplicated().sum())
```

```python
merged_df = merged_df.drop(columns=['User'])
```

```python
#Save merged data set
merged_df.to_csv("merged_dataset.csv", index=False)
```
